###### Department wise codekata performence (pie chart)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for m in data.Department.unique():
    print(m)
    dept = np.where(data['Department'] == m)
    dept_data = data.loc[dept]
    print(dept_data['CodeKata Score'].mean())
```

```python
c = data.groupby('Department')['CodeKata Score'].mean()
```

```python
c
```

```python
plt.pie(c,labels =c.index,autopct = ("%.2f%%"),shadow = True, startangle = 90);
```
